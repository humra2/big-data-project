```python
from pyspark import SparkConf, SparkContext

if __name__ == "__main__":
    # Initialize a SparkSession
    conf = SparkConf().setAppName("WordCount")
    sc = SparkContext(conf=conf)

    # Load the text file into an RDD
    inputFile = sc.textFile("/content/sample3.txt")

    # Tokenize the lines of the file into words
    words = inputFile.flatMap(lambda line: line.split(" "))

    # Count the occurrences of each word
    wordCounts = words.map(lambda word: (word, 1)).reduceByKey(lambda a, b: a + b)

    # Display or save the word count results
    for word, count in wordCounts.collect():
        print(f"{word}: {count}")

    # Stop the SparkContext
    sc.stop()
```

```python
from pyspark import SparkConf, SparkContext

if __name__ == "__main__":
    # Initialize a SparkSession
    conf = SparkConf().setAppName("AlphaCount")
    sc = SparkContext(conf=conf)

    # Load the text file into an RDD
    inputFile = sc.textFile("/content/sample3.txt")

    # Flatten the lines of the file into characters
    characters = inputFile.flatMap(lambda line: list(line))

    # Filter out non-alphabetic characters
    alphabets = characters.filter(lambda char: char.isalpha())

    # Count the occurrences of each alphabetic character
    alpha_counts = alphabets.map(lambda char: (char, 1)).reduceByKey(lambda a, b: a + b)

    # Display or save the character count results
    for alpha, count in alpha_counts.collect():
        print(f"{alpha}: {count}")

    # Stop the SparkContext
    sc.stop()
```
